```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA','JAPAN','MEXICO','COLOMBIA','GERMANY']
population = [1000,800,900,1200,300]
```

```python
plt.bar(country,population,width=0.5,color=['green','purple'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'),rotation=45)
plt.title("Pobalcion por pa√≠s")
plt.xlabel("Pa√≠s")
plt.ylabel("Poblacion")
plt.show()
```

```python
plt.barh(country,population,color=['green','purple'])
plt.show()
```
