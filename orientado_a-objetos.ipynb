```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()   # Creamos una figura (canvas vacío) "Más personalizable"
axes = fig.add_axes([0.1,0.1,0.5,0.9])   #  Añadimos los ejes 
#0,1 primeros=posicion to figure, 2,3=size del lienzo
axes.plot(x,'g--')     # 'b' hace referencia al color azul (blue)
fig.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.9]) # eje principal
axes2 = fig.add_axes([0.2,0.55,0.4,0.3]) # eje interior

# Figura principal axes
axes.plot(x,y,'g--')

# Inserta figura axes2
axes2.plot(y,x,'r--')
axes2.set_facecolor('gray')

fig.show()
```
