```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
# Se usa de forma similar a plt.figure() excepto que devuelve al mismo tiempo un objeto figura y un objeto (o un arreglo) de ejes
fig, axes = plt.subplots()

# Ahora se usa el eje, como antes, para graficar en él
axes.plot(x,y,'g--')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'g--')
axes[1].plot(y,x,'y--')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'g--')
ax2.plot(y,x,'y--')
plt.show()
```

```python
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x,np.cos(x), "b--")
axes[0,1].plot(x,np.sin(x),'y--')
axes[0,2].plot(x,np.tan(x),'g--')
axes[0,3].plot(x,np.cos(x)**2, "r--")
plt.show()
```

```python
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x,np.cos(x), "b--")
ax2.plot(x,np.sin(x),'y--')
ax3.plot(x,np.tan(x),'g--')
ax4.plot(x,np.cos(x)**2, "r--")
fig.tight_layout()    # Para evitar superposición de elementos
```
