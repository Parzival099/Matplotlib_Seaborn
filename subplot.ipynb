```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x,y,'g*-')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)       # 1 fila, dos columnas y se esta referenciando el n√∫mero 1
plt.plot(x,y,'g--')
plt.plot(y,x,'b--')       # Con la referencia a√∫n al plot 1, se dibuja otra g≈ïafica
plt.subplot(1,2,2)       # ahora se referencia el plot n√∫mero 2
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'g--')
plt.plot(y,x,'b--')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
